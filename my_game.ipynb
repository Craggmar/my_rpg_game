```python
if __name__ =='__main__':
    import random
    
attributes_dict = {'s':'strenght', 'd': 'dexterity', 'i':'intelligence', 'e':'stamina'} #słowniczek

equipped_inventory = [] #
current_loot_list=[] #itemy odpowiadające lvlowi
loot_list=[] #wszystkie itemy w grze
```

```python
class Hero():
    
    
    def __init__(self):
         
        self.name= 'Hero'
        #STATS
        #self.strenght = self.strenght()
        self.strenght = 1 #pure dmg, req for items
        self.dexterity = 1 #chance for dodge
        self.intelligence = 1 #spell dmg, mana
        self.stamina = 1 #hp
        
        self.max_health = 10
        self.max_mana = 10
        self.current_health=self.max_health
        self.current_mana=self.max_mana
        self.level = 1
        self.experience = 0
        self.max_experience = 10
        self.attributes_stats = {'s':self.strenght, 'd':self.dexterity, 'i':self.intelligence, 'e':self.stamina}
        self.weapon = 0
        self.armor = 0
        
        update_loot_list(self.level)
    
    def stats_list(self):
        self.stats = [self.strenght, self.dexterity, self.intelligence, self.stamina]
    
    def update_stats(self):
        
        self.strenght=self.attributes_stats['s']
        self.dexterity=self.attributes_stats['d']
        self.intelligence=self.attributes_stats['i']
        self.stamina=self.attributes_stats['e']
            
    def update_bars(self):
        #HP
        self.max_health= 10 + self.stamina*2 -2
        
        #MANA
        self.max_mana = 10 + self.intelligence*2 - 2
        
        
        #EXP
        
        self.max_experience +=10
        
        
    def attack(self):
        return self.strenght+self.weapon
    
    def check_promotion(self):
        if self.experience>=self.max_experience:
            self.level+=1 #lvl+1
            update_loot_list(self.level)
            print(f'Your hero has been promoted! Your hp & mana is full now. Current level: {self.level}. Good job!')
            self.experience=self.experience-self.max_experience #calculate current experience
            self.add_points()
            self.update_stats()
            self.update_bars()
            self.current_health= self.max_health #get max hp
            self.current_mana= self.max_mana #get max mana
            
                  
        
        
    def death(self):
        pass

    
    def add_points(self):
        while True:    
            global att
            att=input('You have 1 xp to spend. Choose an attribute:\n[S]trenght\n[D]exterity]\n[I]ntelligence\n[E] Stamina\n')
            if att.lower() in attributes_dict.keys():
                confirm=input(f'You picked {attributes_dict[att]}. Are you sure? Enter [y] to confirm or any button to cancel')
                if confirm.lower()=='y':
                    self.attributes_stats[att]+=1
                    print(f'Your have now {self.attributes_stats[att]} {attributes_dict[att]}!')
                    break
            else: 
                print('Wrong letter!')
                
            
            
        
        
    def __str__(self):
        return(f'Your stats:\nSTR = {self.strenght}\nDEX = {self.dexterity}\nINT = {self.intelligence}\nSTA = {self.stamina}\nHealth = {self.current_health}\nMana = {self.current_mana}\nHealth = {self.max_health}\nMana = {self.max_mana}\nlvl={self.level}\nXp={self.experience}') 
```

```python
class Spell():
    def __init__(self, name, damage, mana_cost):
        self.name= name
        self.damage = damage
        self.mana_cost = mana_cost
    
    def __str__(self):
        return ('Deal {} dmg + your intelligence. Cost {} mp.'.format(self.damage, self.mana_cost))

class Item():
    def __init__(self, name, itype, ilvl=0, damage=0, armor=0, req_stats=[0,0,0,0]):
        self.name = name
        self.type = itype
        self.ilvl = ilvl
        self.damage = damage
        self.armor = armor
        self.required_stats = req_stats #wymagania przedmiotów
            
        
    def __name__(self):
        return self.name
        
    def __str__(self):
        if self.type=='weapon':
            return f'{self.name}, damage = {self.damage}'
        elif self.type=='armor':
            return f'{self.name}, armor = {self.armor}'
        
        
```

```python
class Monster():
    def __init__(self, health=1, damage=1, mana=0, weapon=0, armor=0):
        self.name = 'Monster'
        self.current_health = health
        self.current_mana = mana
        self.strenght = damage
        self.xp = (self.current_health+self.strenght)/2
        self.armor = armor
        
        
    def attack(self, weapon=0, armor=0):
        return self.strenght+weapon
    
        
    def death(self):
        print("You killed an enemy! You get {} EXP!".format(self.xp))
        if check_loot(): #get loot
            add_loot()
        hero.experience += self.xp#get xp
        hero.check_promotion()
        
    def __str__(self):
        return(f'You met a monster!\nDamage = {self.strenght}\nHealth = {self.current_health}')#\nEXP = {self.exp}')
```

```python
#exploration

def choose_exploration():
    while True:
        find = input('What do u want to do? Press:\n[E] to explore\n[I] to change eq')
        if find.lower() == 'e':
            print(roll_monster(), '\n')
            while True:
                find_again = input('Wanna find a new enemy? Press [y] or [n]: ')
                if find_again.lower() == 'y':
                    print(roll_monster(), '\n')
                    continue
                elif find_again.lower() == 'n':
                    print('You have chosen the enemy!')
                    choose_action()
                else: 
                    print('Wrong! Enter correct letter')
                    continue
                break
            break
        elif find.lower() =='i':
            eq_items()
            break
        else:
            print('Wrong! Enter correct letter')

def eq_items():
    print('Current eqquiped items:', [item.name for item in equipped_inventory])
    print('Your inventory:')
    i=0
    for item in inventory:
        i+=1
        print('['+str(i)+f'] {item.name}')
    print('[0] to exit')
    while True:
        try:
            choose = int(input('What item do u want to use? Pick a number: '))
        except ValueError:
            print('This is not the number!')
            continue
        else:
            if choose==0:
                pass
            elif choose<=i and choose>0 and type(choose)==int:
                item_to_eq=inventory.pop(choose-1)
                hero.weapon=item_to_eq.damage
                hero.armor=item_to_eq.armor
                for item in equipped_inventory:
                    if item.type ==item_to_eq.type:
                        inventory.append(item)
                        equipped_inventory.remove(item)
                equipped_inventory.append(item_to_eq)
                print(f'You wield now', item_to_eq)
                break
            else:
                print('Wrong number!')
                continue
        break
    
        


def roll_monster(lvl=1):
    global monster 
    monster = Monster(random.randint(3*lvl,5*lvl), random.randint(1*lvl,2*lvl))
    return monster


#actions while fight
def choose_action():
    while monster.current_health> 0 and hero.current_health>0:
        print('\nYour hp: {}'.format(hero.current_health))
        print('Your mp: {}'.format(hero.current_mana))
        print("Enemy's hp: {}".format(monster.current_health))
        action = input('What do u want to do now? Press:\n[F]ight\n[D]rink\n[R]un away\n[J]ebnij z fireballa\n')
        if action.lower()=='f':
            attack(hero,monster) #hero attack
            if monster.current_health<=0:
                monster.death()
                continue
            elif check_dodge(hero): #check for dodge
                print ("You dodged an attack!")
            else:
                attack(monster,hero) #monster attack
        elif action.lower()== 'j':
            if check_spell_cost(hero, fireball):
                cast_spell(hero,monster, fireball)
            else:
                continue
            if monster.current_health<=0:
                monster.death()
                continue
            elif check_dodge(hero): 
                print ("You dodged an attack!")
            else:
                 attack(monster,hero)
        elif action.lower()=='d':
            choose_potion()
            if check_dodge(hero): 
                print ("You dodged an attack!")
            else:
                attack(monster,hero) #monster attack
        elif action.lower()=='r':
            attack(monster,hero) #monster attack
            run_away()
            break
        else:
            print('Wrong! Enter correct letter')
            
    if hero.current_health<=0: #do napisania!!!!
        hero.death()

        
#attack

def attack(dealer, victim):
        victim.current_health-=dealer.attack()-victim.armor
        print('{} hit {} for {} damage!'.format(dealer.name, victim.name, dealer.attack()))
#SPELLS        
def cast_spell(dealer, victim, spell):
        damage=dealer.intelligence+spell.damage
        victim.current_health-=damage
        print(f'{dealer.name} casted {spell.name} at {victim.name} for {damage} damage! Ale urwał!\n')
        
        
#POTIONS   
def choose_potion():
    while True:
        potion= input('Choose a potion:\n[H]ealth potion\n[M]ana potion')
        if potion.lower()=='h':
            drink_hp_potion()
            break
        elif potion.lower()=='m':
            drink_mana_potion()
            break
        else:
            print('Wrong! Enter correct letter')
            
def drink_hp_potion():
    if hero.max_health - hero.current_health<5:
        hero.current_health = hero.max_health
    else:
        hero.current_health +=5
    print('You drank hp potion! Your current hp is {}'.format(hero.current_health))

def drink_mana_potion():
    if hero.max_mana - hero.current_mana<5:
        hero.current_mana = hero.max_mana
    else:
        hero.current_mana +=5
    print('You drank mana potion! Your current mana is {}'.format(hero.current_mana))

def run_away():
    print('You have fleed!')
    
def check_loot():
    return random.randint(0,100) <= 20

def add_loot():
    looted_item = random.choice(loot_list)
    print(f'You found a new item: {looted_item}!')
    inventory.append(looted_item)
    
def check_it_req(item):
    hero.stats_list()
    i=0
    for stat in item.req_stats:
        
        if stat > hero.stats_list[]:
            print(f'Too low {stat} !')
            return True
            break
        else:
            return True
        i+=1
            
        
        
    
def check_dodge(character):
    return random.randint(0,100) <= character.dexterity*10

def check_spell_cost(character, spell):
    if spell.mana_cost<=character.current_mana:
        character.current_mana-=spell.mana_cost
        return True
    else:
        print('Not enough mana!')
        return False

def update_loot_list(lvl):
    global loot_list
    for item in loot_list:
        if item.ilvl==lvl:
            current_loot_list.append(item)
```
